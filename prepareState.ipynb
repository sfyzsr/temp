```python
import pandas as pd
df = pd.read_feather("./hs300/sh.600000.feather")
df
```

```python
import matplotlib.dates as mdates
df['date_num'] = mdates.date2num(df['date'])
df
```

```python
df.info()
```

```python
import numpy as np
df['close'] = df['close'].astype('float32')
df['norm_close'] = np.log(df['close'])
df
```

```python
df['60dayEWM'] = df['norm_close'].ewm(span=60, adjust=False).mean()
df
```

```python
import matplotlib.pyplot as plt

#plot sales and 4-day exponentially weighted moving average 
plt.plot(df['norm_close'], label='norm_close')
plt.plot(df['60dayEWM'], label='60dayEWM')
```

```python
df['return'] = df['close'].pct_change()
df['log_return'] = np.log(df['close']).diff()
df
```

```python
# annual return = ((1+r1)*(1+r2)*...*(1+rn))^(1/n) - 1
# = (1 + cumulative return) ^(365/days held) - 1

df['cumulative_return'] = (1 + df['return']).cumprod()

n = 365  # Assuming daily returns
df['annual_return'] = df['cumulative_return'] ** (n / len(df)) - 1
df
```
