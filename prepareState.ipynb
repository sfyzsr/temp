```python
import pandas as pd
df = pd.read_feather("./hs300/sh.600000.feather")
df
```

```python
import matplotlib.dates as mdates
df['date_num'] = mdates.date2num(df['date'])
df
```

```python
df.info()
```

```python
import numpy as np
# log normalized close
df['close'] = df['close'].astype('float32')
df['norm_close'] = np.log(df['close'])
df
```

```python
# now prepare the normalized return for the past month, 2-month, 3-month and 1-year periods
df['60dayEWM'] = df['close'].ewm(span=60, adjust=False).mean()
# get standard derivation by directly call the std()
df
```

```python
import matplotlib.pyplot as plt

#plot sales and 4-day exponentially weighted moving average 
plt.plot(df['norm_close'], label='norm_close')
plt.plot(df['60dayEWM'], label='60dayEWM')
```

```python
df['return'] = df['close'].pct_change()
df['log_return'] = np.log(df['close']).diff()
df
```

```python
# annual return = ((1+r1)*(1+r2)*...*(1+rn))^(1/n) - 1
# = (1 + cumulative return) ^(365/days held) - 1

df['cumulative_return'] = (1 + df['return']).cumprod()

n = 365  # Assuming daily returns
df['annual_return'] = df['cumulative_return'] ** (n / len(df)) - 1
df
```

```python
n = 30  
df['monthly_return'] = df['cumulative_return'] ** (n / len(df)) - 1
n = 60  
df['2monthly_return'] = df['cumulative_return'] ** (n / len(df)) - 1
n = 90  
df['3monthly_return'] = df['cumulative_return'] ** (n / len(df)) - 1
```

```python
df['ema_8']= df['close'].ewm(span=8, adjust=False).mean()
df['ema_16'] = df['close'].ewm(span=16, adjust=False).mean()
df['ema_32'] = df['close'].ewm(span=32, adjust=False).mean()

df['ema_24'] = df['close'].ewm(span=24, adjust=False).mean()
df['ema_48'] = df['close'].ewm(span=48, adjust=False).mean()
df['ema_96'] = df['close'].ewm(span=96, adjust=False).mean()
```

```python
# Calculate the rolling standard deviation
df['rolling_std'] = df['close'].rolling(window=63).std()
df
```

```python
df['price_change'] = df['close'].diff()
period = 30
df['gain'] = df['price_change'].apply(lambda x: x if x > 0 else 0)
df['loss'] = df['price_change'].apply(lambda x: abs(x) if x < 0 else 0)
df['avg_gain'] = df['gain'].rolling(window=period).mean()
df['avg_loss'] = df['loss'].rolling(window=period).mean()
df['rs'] = df['avg_gain'] / df['avg_loss']
df['rsi'] = 100 - (100 / (1 + df['rs']))
df
```

```python
window = 60
df['ex_ante_volatility'] = df['log_return'].ewm(span=window).std()
df
```

```python
df.to_feather("./preprocess/prepared_sh.600000.feather")
```
